The cloud model
===============
To specify the scheduling problem, we first define the model that consists of:
    
- $servers$, a set of active servers constituting the cloud, cardinality $n$
- $VMs$, a set of VMs provided to the users, cardinality $m$

Both sets can change in any moment by creating/deleting VMs or adding/removing servers. These events are not controlled by the scheduler.

The cloud provider defines a set of $r$ resource types reserved for a VM that have to be specified through an SLA. The exact values specified through individual SLAs for these resources are an ordered r-tuple of numbers, defined as a VM's $spec$:
    
$spec_{vm} = \left(res_{vm,i},\, \forall i \in \{1, \ldots, r\}\right),\, \forall vm \in VMs$


Similarly, the capacity of a server is defined as an r-tuple of the resource amounts it has:

$spec_{s} = \left(cap_{s,i},\, \forall i \in \{1, \ldots, r\}\right),\, \forall s \in servers$

In every moment, every VM has to be allocated to some server (belong to its $alloc$ set) that acts as its host. This is the first problem constraint (C1):

$\left(\forall s \in servers \to alloc_s \subseteq VMs \right)\, s.t. \left( \forall vm \in VMs,\, \exists_{=1} s \in servers \to vm \in alloc_s \right)$

The other constraint (C2) is that at any given time a server cannot host VMs that require more resources in sum that it can provide.

$\forall s \in servers,\, \forall i \in \{1, \ldots, r\}  \to  \sum_{vm \in alloc_s} res_{vm,i} < cap_i$

The power consumption model
===========================

Linearly proportional to the number of CPUs.

Supose we have a $vm \in VMs, s \in servers$ where $vm \in alloc_s$. The resource ratio $rr_{vm}$ is the fraction of CPUs allocated to $vm$

$rr = \frac{cores_{vm}}{cores_{s}}$

given that $cores$ represents the number of CPUs in the corresponding $spec$ set.

Roulette wheel selection
-------------------------

```python
class Unit(object):
    def __init__(self, name):
        self.name = name
    def __repr__(self):
        return self.name + ': ' + str(self.fitness)

population = [Unit(name) for name in list('abcdefghijkl')]
for i, unit in enumerate(population):
    unit.fitness = (i+1)/float(len(population))
    unit.rfitness = 1 - unit.fitness
print(population)
```

```python
import random
def selRoulette(individuals, k):
    """Select *k* individuals from the input *individuals* using *k*
    spins of a roulette. The selection is made by at the rfitness attributes,
    assuming that rfitness approaches 1.0 for the best units and 0.0 for the worst.
    
    @param individuals: A list of individuals to select from with rfitness attributes
    @param k: The number of individuals to select.
    @returns: A list of selected individuals.
    
    This function uses the :func:`~random.random` function from the python base
    :mod:`random` module.
    
    .. warning::
       The roulette selection by definition cannot be used for minimization 
       or when the fitness can be smaller or equal to 0.
    """
    s_inds = sorted(individuals, key=lambda u : u.rfitness, reverse=True)
    sum_fits = sum(ind.rfitness for ind in individuals)
    
    chosen = []
    for i in range(k):
        u = random.random() * sum_fits
        sum_ = 0
        for ind in s_inds:
            sum_ += ind.rfitness
            if sum_ > u:
                chosen.append(ind)
                break
    
    return chosen

selRoulette(population, 10)
```

```python
# read geotemporal inputs (warning: based on the current config, not tied to the serialised results!)
import philharmonic as ph
ph._setup()
from philharmonic.simulator.inputgen import el_prices_from_conf, temperature_from_conf
el_price = el_prices_from_conf()
temperature = temperature_from_conf()

def parse_actions(location):
    actions_list, method_names = [], []
    cloud = None
    for measurement_folder in os.listdir(location):
        cloud =  pd.read_pickle(os.path.join(location, 'servers.pkl'))
        if os.path.isdir(os.path.join(location, measurement_folder)) and measurement_folder != 'unused':
            name = measurement_folder.replace('_', ' ').upper()
            actions = pd.read_pickle(os.path.join(location, measurement_folder, 'schedule.pkl'))
            
            #actions.sort()
            actions_list.append(actions)
            method_names.append(name)
    return actions_list, method_names, cloud
#actions_list, method_names, cloud = parse_actions(os.path.join(RES_LOC, 'usa-3month/world-bfd-initial'))
actions_list, method_names, cloud = parse_actions(os.path.join(RES_LOC, 'usa-3month/test'))

# TODO: also mark unscheduled VMs and overcapacitated PMs, add geotemporal inputs to the graph
#schedule
#cloud.servers
transparent_legend = lambda ax : ax.get_legend().get_frame().set_alpha(0.7)

def calculate_location_vms(cloud, actions):
    cloud.reset_to_initial()
    locations = list(set([s.loc for s in cloud.servers]))
    location_vms = []
    location_pms = []
    unallocated_vms = []
    overcapacitated_pms = []

    unique_actions_index = actions.index.unique()
    for t in unique_actions_index:
        #print("{} / {}".format(t, len(unique_actions_index)))
        # TODO: precise indexing, not dict
        if isinstance(actions[t], pd.Series):
            for action in actions[t].values:
                cloud.apply(action)
        else:
            action = actions[t]
            cloud.apply(action)
        state = cloud.get_current()
        # count VMs at location
        location_vms_current = {loc: 0 for loc in locations}
        for server in state.alloc: # calculate the current number of VMs at location
            location_vms_current[server.loc] += len(state.alloc[server])
        location_vms.append(location_vms_current)
        # count active PMs at location
        location_pms_current = {loc: 0 for loc in locations}
        active_pms = [pm for pm in state.alloc if len(state.alloc[pm]) > 0]
        for server in active_pms:
            location_pms_current[server.loc] += 1
        location_pms.append(location_pms_current)
        # unallocated VMs
        unallocated_vms.append(len(state.unallocated_vms()))
        # overcapacitated PMs
        overcapacitated_pms.append(len(state.overcapacitated_servers()))
    location_vms = pd.DataFrame(location_vms, actions.index.unique())
    location_pms = pd.DataFrame(location_pms, actions.index.unique())
    unallocated_vms = pd.Series(unallocated_vms, actions.index.unique())
    overcapacitated_pms = pd.Series(overcapacitated_pms, actions.index.unique())
    
    #start, end = actions.index[0] - pd.offsets.Hour(1), actions.index[-1] + pd.offsets.Hour(2)
    start, end = actions.index[0], actions.index[-1]
    
    return location_vms, location_pms, unallocated_vms, overcapacitated_pms, start, end

def draw_scheduling_plots(location_vms, location_pms, unallocated_vms, overcapacitated_pms, title=None):
    figure(figsize=(20, 4))
    num_subplots = 3
    
    subplot(1, num_subplots, 1)
    location_vms.plot(ax=plt.gca()); transparent_legend(plt.gca())
    plt.title('VMs')
    ymin, ymax = plt.gca().get_ylim()
    plt.gca().set_ylim(ymin, ymax * 1.1)
    
    subplot(1, num_subplots, 2)
    location_pms.plot(ax=plt.gca()); transparent_legend(plt.gca())
    plt.title('active PMs')
    ymin, ymax = plt.gca().get_ylim()
    plt.gca().set_ylim(ymin, ymax * 1.1)
    
    subplot(1, num_subplots, 3)
    unallocated_vms.plot(ax=plt.gca(), label='unallocated VMs', style='k-')
    overcapacitated_pms.plot(ax=plt.gca(), label='overcapacitated PMs', style='r--')
    ymin, ymax = plt.gca().get_ylim()
    plt.gca().set_ylim(ymin, ymax * 1.1)
    legend(); transparent_legend(plt.gca())
    plt.title('Constraint violations')
    
    if title:
        plt.suptitle(title)
    savefig(imgloc('active_machines-{}.pdf'.format(title)), bbox_inches='tight')
    show()

def visualise_actions(location):
    print(location)
    actions_list, method_names, cloud = parse_actions(os.path.join(RES_LOC, location))
    for actions, method_name in zip(actions_list, method_names):
        print(method_name)
        location_vms, location_pms, unallocated_vms, overcapacitated_pms, \
        start, end = calculate_location_vms(cloud, actions)
        draw_scheduling_plots(location_vms, location_pms, unallocated_vms, overcapacitated_pms, title=method_name)
        
    figure(figsize=(22, 4))
    subplot(1, 2, 1)
    el_price[start:end].plot(ax=plt.gca()); transparent_legend(plt.gca())
    subplot(1, 2, 2)
    temperature[start:end].plot(ax=plt.gca()); transparent_legend(plt.gca())
    savefig(imgloc('active_machines-geotemporal_inputs.pdf'))
#visualise_actions('usa-3month/world-bfd-initial')
#visualise_actions('usa-3month/world-realtemp')
visualise_actions('usa-3month/test')
```

```python
figsize = (6, 1.5)
def plot_migrations_histogram(migrations_count):
    #plt.figure(figsize=(4.5, 1.5))
    plt.figure(figsize=figsize)
    plt.subplot(121)
    migrations_count.hist(hatch='/', color='w', grid=False)
    plt.locator_params(axis='y', nbins=4)
    #xlim(0, 5)
    ylabel('hours')
    xlabel('migration rate (#/h)')
    plt.subplot(122)
    migrations_count[migrations_count>0].hist(hatch='/', color='w', grid=False, bins=7)
    plt.locator_params(axis='y', nbins=4)
    #xlim(1, 8)
    xlabel('migration rate (#/h)')
    ymin, ymax = plt.gca().get_ylim()
    plt.ylim(ymin, ymax * 1.1)
    plt.savefig(loc('migrations-rate.pdf'), bbox_inches='tight')
    plt.show()
    
def plot_aggregate_migrations(aggr_migrations, aggregation_interval = '24h'):
    #plt.figure(figsize=(4, 1.2))
    plt.locator_params(nbins=4)
    aggr_migrations.plot(color='r', figsize=figsize, lw=1.5)
    xlabel('')
    ylabel('rate (#/day)')
    plt.savefig(loc('migrations-aggregated.pdf'), bbox_inches="tight")
    plt.show()

#from scikits import bootstrap
import scikits.bootstrap as bootstrap

def visualise_migration_frequency(schedule_path):
    # parse
    schedule = pd.read_pickle(schedule_path)

    # filter migrations
    is_migration = lambda a : a.name == 'migrate'
    migrations = schedule[schedule.apply(is_migration)]
    #print(migrations.head(3))

    # plot straight
    migrations_count = migrations.resample('1H', how='count')
    plot_migrations_histogram(migrations_count)
    
    # separate per VM (MultiIndex DataFrame approach)
    times = np.array(schedule.index)
    vms = np.array(schedule.apply(lambda action : action.vm))
    vm_schedule = pd.Series(schedule.values, index=[times, vms])
    vm_schedule.index.names = ['time', 'vm']
    migrations = vm_schedule[vm_schedule.apply(is_migration)]
    aggregation_interval = '24h'
    # drop rows with duplicate indices
    # http://stackoverflow.com/questions/13035764/
    migrations = migrations.reset_index().drop_duplicates(['time', 'vm']).set_index(['time', 'vm'])[0]
    time_aggr = migrations.unstack().resample(aggregation_interval, how='count')
    aggr_migrations = time_aggr.transpose().max()
    plot_aggregate_migrations(aggr_migrations, aggregation_interval)
    return aggr_migrations
    
#schedule_path = loc(os.path.join('results', 'usa-3month/test', 'ga_hybrid', 'schedule.pkl'))
#! ls $schedule_path
#visualise_migration_frequency(schedule_path)

schedule_path = loc(os.path.join('results', 'usa-3month/large_scale2', 'ga_hybrid', 'schedule.pkl'))
aggr_migrations = visualise_migration_frequency(schedule_path)
```

Migration overhead

```python
alpha = 0.512
beta = 20.165
E_mig = lambda V_mig : alpha*V_mig + beta
V_mig = lambda V_mem, R, D, n : V_mem * (1-(D/float(R))**(n+1))/(1-D/float(R))
T_mig = lambda V_mig, R : V_mig/(R/8.) # R assumed to be in Mb/s
# example
R, D = 300, 100
V_mem = 1000 # MB
V_thd = 100 # treshold after which post-copying starts
n = math.ceil(math.log(V_thd/float(V_mem), D/float(R)))
#E_mig(V_mig(4, 3, 1, 2))
E_mig(V_mig(V_mem, R, D, n))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.path import Path
from matplotlib.spines import Spine
from matplotlib.projections.polar import PolarAxes
from matplotlib.projections import register_projection

def _radar_factory(num_vars):
    theta = 2*np.pi * np.linspace(0, 1-1./num_vars, num_vars)
    theta += np.pi/2

    def unit_poly_verts(theta):
        x0, y0, r = [0.5] * 3
        verts = [(r*np.cos(t) + x0, r*np.sin(t) + y0) for t in theta]
        return verts

    class RadarAxes(PolarAxes):
        name = 'radar'
        RESOLUTION = 1

        def fill(self, *args, **kwargs):
            closed = kwargs.pop('closed', True)
            return super(RadarAxes, self).fill(closed=closed, *args, **kwargs)

        def plot(self, *args, **kwargs):
            lines = super(RadarAxes, self).plot(*args, **kwargs)
            for line in lines:
                self._close_line(line)

        def _close_line(self, line):
            x, y = line.get_data()
            # FIXME: markers at x[0], y[0] get doubled-up
            if x[0] != x[-1]:
                x = np.concatenate((x, [x[0]]))
                y = np.concatenate((y, [y[0]]))
                line.set_data(x, y)

        def set_varlabels(self, labels):
            self.set_thetagrids(theta * 180/np.pi, labels)

        def _gen_axes_patch(self):
            verts = unit_poly_verts(theta)
            return plt.Polygon(verts, closed=True, edgecolor='k')

        def _gen_axes_spines(self):
            spine_type = 'circle'
            verts = unit_poly_verts(theta)
            verts.append(verts[0])
            path = Path(verts)
            spine = Spine(self, spine_type, path)
            spine.set_transform(self.transAxes)
            return {'polar': spine}

    register_projection(RadarAxes)
    return theta

def radar_graph(labels, combinations, colors):
    N = len(labels) 
    theta = _radar_factory(N)
    #max_val = max(max(combinations[0]), max(values))
    max_val = 1.
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1, projection='radar')
    for combination, color in zip(combinations, colors):
        ax.plot(theta, combination, color=color)
    ax.set_varlabels(labels)
    plt.show()
    #plt.savefig("radar.png", dpi=100)

labels = ['w_util', 'w_cost', 'w_sla', 'w_constraint']
values1 = [0.3, 0.7, 0.0, 0.0]
values2 = [0.0, 0.1, 0.3, 0.6]
#optimum = [5, 3, 2, 4]
 
radar_graph(labels, [v1, v2, v3, v4, v5], colors=['r', 'orange', 'y', 'g', 'b'])
```
